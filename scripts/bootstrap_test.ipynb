Test 1
------

Forward model using same method followed by inversing the model.  Expect to get the same answer out as we put in.  In this test we do not scale any of the SIOPs, so we expect to see all 1s returned

$A_{\lambda_m,n} =
 \begin{pmatrix}
  a_{\lambda 1,\phi} & a_{\lambda 1,m} & a_{\lambda 1,d} & a_{\lambda 1,g} \\
  a_{\lambda 2,\phi} & a_{\lambda 2,m} & a_{\lambda 2,d} & a_{\lambda 2,g}  \\
  \vdots  & \vdots  & \vdots & \vdots \\
  a_{\lambda_m,\phi} & a_{\lambda_m,m} & a_{\lambda_m,d} & a_{\lambda_m,g}
 \end{pmatrix}$
 
  $b_{\lambda_m} = \begin{pmatrix}
 b_{\lambda 1, k} \\
 b_{\lambda 2, k} \\
 \vdots \\
 b_{\lambda m, k}
 \end{pmatrix} $
 
 $x = \begin{pmatrix}
 \phi \\
 m \\
 d \\
 g \\
 \end{pmatrix} $
 
 Where
 
 $A(\lambda) = \left(a_\phi\frac{b_b}{a} - b_\phi \right)\phi + \left(a_m\frac{b_b}{a} - b_m \right)m + \left(a_d\frac{b_b}{a} - b_d \right)d + \left(a_g\frac{b_b}{a} - b_g \right)g$
 
 And 
 
 $b(\lambda) = -\left(a_w\frac{b_b}{a} - b_w \right)$

The solution of X is the solution to the Matrix equation 

$Ax=b$

```python
def run_bootstrap(a_phi, a_m, a_d, a_g, a_w, bb_phi, bb_m, bb_d, bb_g, bb_w, phi=1, m=1, d=1, g=1, num_bootstraps=1, noise=0):
    Aa = np.zeros((wavelength.shape[0], 4))
    Aa[:, 0] = a_phi.T #* phi
    Aa[:, 1] = a_m.T #* m
    Aa[:, 2] = a_d.T #* d
    Aa[:, 3] = a_g.T #* g

    Ab = np.zeros((wavelength.shape[0], 4))
    Ab[:, 0] = bb_phi.T #* phi
    Ab[:, 1] = bb_m.T #* m
    Ab[:, 2] = bb_d.T #* d
    Ab[:, 3] = bb_g.T #* g

    A = np.zeros((Aa.shape[0], Aa.shape[1]))
    
    a = a_phi + a_d + a_m + a_g + a_w
    bb = bb_phi + bb_m + bb_g + bb_w
    bb_over_a = bb / a 

    scale = np.array([phi, m, d, g])

    for i_iter in range(Aa.shape[1]):
        A[:, i_iter] = (Aa[:, i_iter] * (bb_over_a[:]) - Ab[:, i_iter]) * scale[i_iter]

    b = (scipy.squeeze(a_w.T) * bb_over_a - scipy.squeeze(bb_w.T)) * -1

    #--------------------------------------------------#
    # Now we replicate the matricies and add the noise
    #--------------------------------------------------#
    A = scipy.tile(A, (num_bootstraps, 1))
    b = scipy.tile(b, (num_bootstraps))

    #noise_a = (scipy.random.randn(np.int(A.shape[0]), np.int(A.shape[1])) * noise) + 1
    #noise_b = (scipy.random.randn(np.int(b.shape[0])) * noise) + 1
    noise_a = scipy.random.uniform(low=(1 - noise), high=(1 + noise), size=A.shape)
    noise_b = scipy.random.uniform(low=(1 - noise), high=(1 + noise), size=b.shape)
    
    #import pylab
    #pylab.plot(noise_b, '*')
    #pylab.show()
    #print(noise_b.max())
    
    A = A * noise_a
    b = b * noise_b

    x = scipy.sparse.linalg.lsqr(A, b.T)
    return x
```

$x\begin{pmatrix}
 \phi \\

 g \\
 \end{pmatrix} = $

```python
x = run_bootstrap(a_phi, a_m, a_d, a_g, a_w, bb_phi, bb_m, bb_d, bb_g, bb_w)
for i in x[0]:
    print(i)
```

Test 2

We repeat Test 1 but we make 1000 bootstraps with 0% noise

```python
scale_vals = np.asarray([1, 1, 1, 1])
num_bootstraps = 10000
noise = 0.05

phi = scale_vals[0]
m = scale_vals[1]
d = scale_vals[2]
g = scale_vals[3]

x = run_bootstrap(a_phi, a_m, a_d, a_g, a_w, bb_phi, bb_m, bb_d, bb_g, bb_w,
                  phi=phi, m=m, d=d, g=g,
                  num_bootstraps=num_bootstraps, noise=noise)

residual = ((x[0] - scale_vals) / x[0]) * 100

print('value :: residual %')
for i_iter, i in enumerate(x[0]):
    print(str(i) + ' :: ' + str(residual[i_iter]) + '%' )
    #print(i)
```
