Deconvolution of Specific Inherent Optical Properties (SIOPs)
================

Forward model using same method followed by inversing the model.  Expect to get the same answer out as we put in.  In this test we do not scale any of the SIOPs, so we expect to see all 1s returned

$A_{\lambda_m,n} =
 \begin{pmatrix}
  a_{\lambda_1,\phi} & a_{\lambda_1,m} & a_{\lambda_1,d} & a_{\lambda_1,g} \\
  a_{\lambda_2,\phi} & a_{\lambda_2,m} & a_{\lambda_2,d} & a_{\lambda_2,g}  \\
  \vdots  & \vdots  & \vdots & \vdots \\
  a_{\lambda_m,\phi} & a_{\lambda_m,m} & a_{\lambda_m,d} & a_{\lambda_m,g}
 \end{pmatrix}$  $b_{\lambda_m} = \begin{pmatrix}
 b_{\lambda_1, k} \\
 b_{\lambda_2, k} \\
 \vdots \\
 b_{\lambda_m, k}
 \end{pmatrix} $ $x = \begin{pmatrix}
 \phi \\
 m \\
 d \\
 g \\
 \end{pmatrix} $
 
 Where
 
 $A(\lambda) = \left(a_\phi\frac{b_b}{a} - b_\phi \right)\phi + \left(a_m\frac{b_b}{a} - b_m \right)m + \left(a_d\frac{b_b}{a} - b_d \right)d + \left(a_g\frac{b_b}{a} - b_g \right)g$
 
 And 
 
 $b(\lambda) = -\left(a_w\frac{b_b}{a} - b_w \right)$

The solution of $x$ is the solution to the Matrix equation 

$Ax=b$

Bootstrap method
----------------
The boot strap method is a technique that can increase empirical confidence of estimated parameters.  The method works by simulating noise inherent in the data which has the effect of synthentically increasing the sample size of the training set of data.

One variation of the boot strap method is to define a training set of data and a validation set of data.  The method makes an estimate of the noise and assumes that it is small with respect to the parameters and have zero mean.

$y = f(x) + \epsilon$    

Where $\epsilon$ is noise with an empirical distribution.  A basic summary of the process is as follows.

* For a training set of measurements ${x_i}$ estimate a noise level.
* Resample the measurements by generating a new set of measurements ${x_i + \epsilon_i}$
* Invert the model of parameters of interest
* cross-validate the training set against the validation set of data
* Repeat this procoess for a new choice of noise until there is no improvement in the prediction

The following sections investigates the boot strap model by first testing the affectiveness of the method on a synthetic dataset followed by testing the method on measurements measured in the Ligurian sea and validated with in-situ measurements.

Model
-----

Below the dconvolution via bootstrap technique code in Python.  The function defined is:

```
run_bootstrap(a_phi, a_m, a_d, a_g, a_w, bb_phi, bb_m, bb_d, bb_g, bb_w,
            phi=1, m=1, d=1, g=1, num_bootstraps=1, noise=0)
```

Where:

* a_phi is $a_\phi$: absorption due to phytoplankton
* a_m is $a_m$: absorption due to suspended minerals
* a_d is $a_d$: absorption due to organic detritus
* a_g is $a_g$: absorption due to gelbstoff
* a_w is $a_w$: absorption of pure sea water
* bb_phi is $b_{b\phi}$: backscatter due to phytoplankton
* bb_m is $b_{bm}$: backscatter due to suspended minerals
* bb_d is $b_{bd}$: backscatter due to organic detritus
* bb_g is $b_{bg}$: backscatter due to glebstoff
* bb_w is $b_{bw}$: backscatter due to pure sea water 
* phi is $\phi$: concentration of phytoplankton
* m is $m$: relative concentration of minerals with respect the SIOP
* d is $d$: relative concentration of detritus with respect to the SIOP
* g is $g$: relative concentration of gelbstoff with respect to the SIOP
* num_bootstraps: is the number replication of in input data
* noise is percentage: from mean of 1 standard deviation random noise

Here `(phi, m, d , g)` are 'helper parameters' which make it possible to artificially scale the inputs.  In practice these should be left at 1.0 (i.e. no scaling) as the scaling is what the model is trying to invert.


````python
def run_bootstrap(a_phi, a_m, a_d, a_g, a_w, bb_phi, bb_m, bb_d, bb_g, bb_w, phi=1, m=1, d=1, g=1, num_bootstraps=1, noise=0):
    Aa = np.zeros((wavelength.shape[0], 4))
    Aa[:, 0] = a_phi.T #* phi
    Aa[:, 1] = a_m.T #* m
    Aa[:, 2] = a_d.T #* d
    Aa[:, 3] = a_g.T #* g

    Ab = np.zeros((wavelength.shape[0], 4))
    Ab[:, 0] = bb_phi.T #* phi
    Ab[:, 1] = bb_m.T #* m
    Ab[:, 2] = bb_d.T #* d
    Ab[:, 3] = bb_g.T #* g

    A = np.zeros((Aa.shape[0], Aa.shape[1]))
    
    a = a_phi + a_d + a_m + a_g + a_w
    bb = bb_phi + bb_m + bb_g + bb_w
    bb_over_a = bb / a 

    scale = np.array([phi, m, d, g])

    for i_iter in range(Aa.shape[1]):
        A[:, i_iter] = (Aa[:, i_iter] * (bb_over_a[:]) - Ab[:, i_iter]) * scale[i_iter]

    
    b = (scipy.squeeze(a_w.T) * bb_over_a - scipy.squeeze(bb_w.T)) * -1
    
    #--------------------------------------------------#
    # Now we replicate the matricies and add the noise
    #--------------------------------------------------#
    A = scipy.tile(A, (num_bootstraps, 1))
    b = scipy.tile(b, (num_bootstraps))

    #noise_a = (scipy.random.randn(np.int(A.shape[0]), np.int(A.shape[1])) * noise) + 1
    #noise_b = (scipy.random.randn(np.int(b.shape[0])) * noise) + 1
    noise_a = scipy.random.uniform(low=(1 - noise), high=(1 + noise), size=A.shape)
    noise_b = scipy.random.uniform(low=(1 - noise), high=(1 + noise), size=b.shape)
    
    A = A * noise_a
    b = b * noise_b

    x = scipy.sparse.linalg.lsqr(A, b.T)
    
    return 1.0 / x[0] , x[1]
````

Test 1
------

Here the test is run with 0% noise and no replication of the data.  The expect return values are 1.0 for all values as the model should just return the concentration of the inputs with respect to the inputs.  There is no artificial scalling of the parameters

````python
x = run_bootstrap(a_phi, a_m, a_d, a_g, a_w, bb_phi, bb_m, bb_d, bb_g, bb_w)
for i in x[0]:
    print(i)
````

Test 2
------

Test 1 is repeated with 10000 bootstraps and 0% noise.  The expected return values are also expected to be 1.0 for all values as no noise has been added.  The model is over determined however all new bootstraps are copies of the inputs and there fore linearly dependent.  This test is primarily to test the software and the hardware memory requirements.

````python
scale_vals = np.asarray([1, 1, 1, 1])
num_bootstraps = 10000
noise = 0.0

phi = scale_vals[0]
m = scale_vals[1]
d = scale_vals[2]
g = scale_vals[3]

x = run_bootstrap(a_phi, a_m, a_d, a_g, a_w, bb_phi, bb_m, bb_d, bb_g, bb_w,
                  phi=phi, m=m, d=d, g=g,
                  num_bootstraps=num_bootstraps, noise=noise)

residual = ((x[0] - scale_vals) / x[0]) * 100

print('value    :: residual %')
print('--------------------')
for i_iter, i in enumerate(x[0]):
    print('{0:4f} :: {1:4f} %' ).format(i, residual[i_iter])
    #print(i)
````
